#### Q1. Explain the difference between greedy and non-greedy syntax with visual terms in as few words as possible. What is the bare minimum effort required to transform a greedy pattern into a non-greedy one? What characters or characters can you introduce or change?
**Ans:** In regular expressions, greedy syntax refers to a pattern that tries to match as much of the input string as possible. The opposite of greedy syntax is non-greedy syntax, which tries to match as little of the input string as possible.

To transform a greedy pattern into a non-greedy pattern, you can add a `?` character after a repetition operator (e.g., `*, +, ?, {m,n}`). This changes the behavior of the repetition operator from greedy to non-greedy, so that it tries to match as little of the input string as possible.

For example, consider the pattern .*:

With greedy syntax, this pattern would match the entire input string. For example, when matching against the string abc, the pattern `.*` would match abc.

With non-greedy syntax, this pattern would match as little of the input string as possible. To make the pattern non-greedy, we add a `?` character after the repetition operator: `.*?.` When matching against the string `abc`, the pattern `.*?` would match only the first character `a`.

```python
import re
print(re.findall("a*", "a")) # Greedy Match Synatx
print(re.findall("a*?", "aaaaa")) # Non Greddy Syntax
```

#### Q3. In a simple match of a string, which looks only for one match and does not do any replacement, is the use of a nontagged group likely to make any practical difference?
**Ans:** In this Case The Non Tagged Group will not make any diffrence in this case.

```python
import re
phoneNumRegex = re.compile(r'\d\d\d')
num = phoneNumRegex.search('My number is 234-567-8901.')
print(f'Phone number found -> {num.group()}') # Non Tagged group
print(f'Phone number found -> {num.group(0)}') # Tagged Group
```

#### Q8. Can you identify repeated items within a target string using named groups, as in "The cow jumped over the moon"?

```python
import re
text = "The cow jumped over the moon"
regobj=re.compile(r'(?P<w1>The)',re.I)
regobj.findall(text)
```
